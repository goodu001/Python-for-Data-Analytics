```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'filmlengthsratings.csv'
df = pd.read_csv(file_path, encoding='ISO-8859-1')

# Grouping the data by 'Year' to get the average 'Length_Minutes'
avg_film_length_by_year = df.groupby('Year')['Length_Minutes'].mean().reset_index()

# Creating the plot
plt.figure(figsize=(10,6))
plt.plot(avg_film_length_by_year['Year'], avg_film_length_by_year['Length_Minutes'], marker='o', linestyle='-', color='b')
plt.title('Average Film Length Over Time', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Average Film Length (Minutes)', fontsize=14)
plt.grid(True)
```
